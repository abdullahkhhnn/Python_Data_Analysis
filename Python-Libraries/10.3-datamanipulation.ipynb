```python
## Handling Missing Values
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
### filling missing values with the mean of the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
## renaiming columns
df.rename(columns={'Sale Date':'Sales Date'})
df.head()
```

```python
## change datatypes
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New Value']=df['Value'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating And Grouping
df.head()
```

```python
grouped_mean=df.groupby('Product Category')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product Category', 'Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_mean1=df.groupby(['Product Category','Region'])['Value'].mean()
print(grouped_mean1)
```

```python
## aggregate multiple functions
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
print(grouped_agg)

total_sales=df.groupby('Product Category')['Sales'].sum()
print(total_sales)
```

```python
### Merging and joining Dataframes
# Create sample DataFrames
df1 = pd.DataFrame({"Key": ['A', 'B', 'C'], "Value1": [1, 2, 3]})
df2 = pd.DataFrame({"Key": ['A', 'B', 'D'], "Value2": [4, 5, 6]})


print(df1)
print(df2)
```

```python
## Merge Datafranme on the 'Key Columns'
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
